```python
def categorize_order_pattern(order_frequency):
    if 1 <= order_frequency <= 7:
        return 'Frequent'
    elif 8 <= order_frequency <= 30:
        return 'Regular'
    elif 31 <= order_frequency <= 90:
        return 'Occasional'
    # elif 61 <= order_frequency <= 90:
    #     return 'Occasional'
    else:
        return 'Infrequent'

# Apply the function to create a new column 'ordering_pattern'
merged_df['ordering_pattern'] = merged_df['order_frequency'].apply(categorize_order_pattern)
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from collections import Counter

# Create an imbalanced dataset for illustration
X, y = make_classification(n_samples=100, n_features=2, n_classes=2, n_informative=2,
                           n_redundant=0, n_clusters_per_class=1, weights=[0.9, 0.1], flip_y=0, random_state=42)

# Display the original class distribution
print("Original class distribution:", Counter(y))

# Plot the original dataset
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', marker='o', edgecolors='k')
plt.title("Original Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Apply SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Display the resampled class distribution
print("Resampled class distribution:", Counter(y_resampled))

# Plot the resampled dataset
plt.scatter(X_resampled[:, 0], X_resampled[:, 1], c=y_resampled, cmap='viridis', marker='o', edgecolors='k')
plt.title("SMOTE Resampled Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
